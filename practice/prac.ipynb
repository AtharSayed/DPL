```python
# 1. Implementing Sigmoid Activation by taking input array from numpy array. 

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function using TensorFlow
def sigmoid_tensorflow(x):
    return tf.nn.sigmoid(x)

# Sigmoid activation function default  
# def sigmoid(x):
#    return 1 / (1 + np.exp(-x))

input_values = np.array([-10, -5, -1, 0, 1, 5, 10], dtype=np.float32)

# x = np.linspace(-10, 10, 400)
# y = sigmoid(x)

input_tensor = tf.convert_to_tensor(input_values)

sigmoid_values = sigmoid_tensorflow(input_tensor)

print("Input Values:", input_values)
print("Sigmoid Output Values (TensorFlow):", sigmoid_values.numpy())

x_values = np.linspace(-10, 10, 100, dtype=np.float32)  # Input range from -10 to 10
x_tensor = tf.convert_to_tensor(x_values)
y_values = sigmoid_tensorflow(x_tensor)

plt.figure(figsize=(4, 3))
plt.plot(x_values, y_values.numpy(), label='Sigmoid Function', color='red')
plt.axhline(0, color='black',linewidth=0.5)  # X-axis
plt.axvline(0, color='black',linewidth=0.5)  # Y-axis
plt.title("Sigmoid Activation Function (TensorFlow)")
plt.xlabel("Input Values")
plt.ylabel("Sigmoid Output")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 2. Implementing ReLU activation by taking input array from numpy 


import numpy as np
import matplotlib.pyplot as plt

# Function to apply ReLU activation
def relu(input_array):
    return np.maximum(0, input_array)

# Demo array
input_array = np.array([-5, 3, -1, 7, -2, 4])

# Applying ReLU
output_array = relu(input_array)

# Display the result
print("Input Array: ", input_array)
print("Output Array (ReLU): ", output_array)

# Plotting the input and output arrays
plt.figure(figsize=(5, 3))
plt.plot(input_array, label='Input Array', marker='o', color='b', linestyle='-', markersize=8)
plt.plot(output_array, label='ReLU Output', marker='o', color='r', linestyle='--', markersize=8)

# Adding labels and title (corrected line here)
plt.title('Input Array vs ReLU Output')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# 3. Implementing Softmax activation by taking input array from numpy 


import numpy as np
import matplotlib.pyplot as plt

# Function to apply Softmax activation
def softmax(input_array):
    exp_values = np.exp(input_array - np.max(input_array))  # Subtract max to prevent overflow
    return exp_values / np.sum(exp_values)

# Demo array
input_array = np.array([2.0, 1.0, 0.2, 3.0, 0.5])

# Applying Softmax
output_array = softmax(input_array)

# Display the result
print("Input Array: ", input_array)
print("Output Array (Softmax): ", output_array)

# Plotting the input and output arrays
plt.figure(figsize=(5, 3))
plt.plot(input_array, label='Input Array', marker='o', color='b', linestyle='-', markersize=8)
plt.plot(output_array, label='Softmax Output', marker='o', color='r', linestyle='--', markersize=8)

# Adding labels and title
plt.title('Input Array vs Softmax Output')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# 4. Implementing Tanh activation function by taking input array from numpy .

import numpy as np
import matplotlib.pyplot as plt

# Function to apply tanh activation
def tanh(input_array):
    return np.tanh(input_array)

# Demo array
input_array = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])

# Applying tanh
output_array = tanh(input_array)

# Display the result
print("Input Array: ", input_array)
print("Output Array (tanh): ", output_array)

# Plotting the input and output arrays
plt.figure(figsize=(5, 3))
plt.plot(input_array, label='Input Array', marker='o', color='b', linestyle='-', markersize=8)
plt.plot(output_array, label='tanh Output', marker='o', color='r', linestyle='--', markersize=8)

# Adding labels and title
plt.title('Input Array vs tanh Output')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
