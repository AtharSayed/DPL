```python
# Implementing Sigmoid Activation by taking sample data  from numpy array. 


import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function using TensorFlow
def sigmoid_tensorflow(x):
    return tf.nn.sigmoid(x)

# Sigmoid activation function default  
# def sigmoid(x):
#    return 1 / (1 + np.exp(-x))

input_values = np.array([-10, -5, -1, 0, 1, 5, 10], dtype=np.float32)

# x = np.linspace(-10, 10, 400)
# y = sigmoid(x)

input_tensor = tf.convert_to_tensor(input_values)

sigmoid_values = sigmoid_tensorflow(input_tensor)

print("Input Values:", input_values)
print("Sigmoid Output Values (TensorFlow):", sigmoid_values.numpy())

x_values = np.linspace(-10, 10, 100, dtype=np.float32)  # Input range from -10 to 10
x_tensor = tf.convert_to_tensor(x_values)
y_values = sigmoid_tensorflow(x_tensor)

plt.figure(figsize=(4, 3))
plt.plot(x_values, y_values.numpy(), label='Sigmoid Function', color='red')
plt.axhline(0, color='black',linewidth=0.5)  # X-axis
plt.axvline(0, color='black',linewidth=0.5)  # Y-axis
plt.title("Sigmoid Activation Function (TensorFlow)")
plt.xlabel("Input Values")
plt.ylabel("Sigmoid Output")
plt.grid(True)
plt.legend()
plt.show()
```
